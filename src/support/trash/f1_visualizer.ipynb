```python
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
array = np.array([[0.47], [0.47], [0.47], [0.55], [0.79], [0.84], [0.87]])
column_values = ["F1"]
df = pd.DataFrame(data = array, columns = column_values)
df["Budget"] = ["5", "20", "125", "625", "1250", "3000", "9000"]

df
```

```python
sns.set(style='whitegrid', font_scale = 2.8)

plt.figure(figsize=(21,9))
g = sns.barplot(x="Budget", y="F1", data=df, palette="Blues_d")
#g.set_yscale("log")
#g.tick_params(axis='x', rotation=45)
#change_width(g, .4)

g.set_xticklabels(["5", "20", "125", "625", "1250", "3000", "9000"])
plt.tight_layout()

plt.savefig('/home/scala/f1_vs_nSamples.pdf')
```
